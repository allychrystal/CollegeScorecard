```python
import pandas as pd
school_scores = pd.read_csv('School_Scores')
```

### What is your ideal school based on your financial status, ideal salary and test scores?

What is the maximum tuition you can afford? 

```python
max_tuition = 35000
```

What is your goal salary in 10 years? 

```python
goal_salary = 80000
```

What is your SAT (verbal and critical reading only) or ACT score?

```python
sat_score = 1300
```

```python
act_score = 0
```

What is your family's income bracket? 
Input:  high, average, low (no spaces, no capital letters)

```python
bracket = 'low'
```

# Add ACT score Run Code Below to See Your Results:


```python
score_qual = sat_score >= school_scores['SAT_25th_Cum']
qual_1 = school_scores.loc[score_qual, :]
tuition_qual = goal_salary <= qual_1['Earnings_10yrs']
qual_2 = qual_1.loc[tuition_qual, :]
if bracket == 'low':
    price_qual = max_tuition >= qual_2['Low_Net_Price']
elif bracket == 'high':
    price_qual = max_tuition >= qual_2['High_Net_Price']
else:
    price_qual = max_tuition >= qual_2['Avg_Net_Price']

your_schools = qual_2.loc[price_qual,:]
your_schools
```

```python
print('Congratulations! {} schools match your criteria.'.format(your_schools.shape[0]))
```
